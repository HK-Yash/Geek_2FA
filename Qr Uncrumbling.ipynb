```python
import numpy as np
import Shuffling as sf
import cv2 
import matplotlib.pyplot as plt
```

```python
def get_data(qr_name):
    with open("Data.txt",'r') as file:
        lines = file.readlines()
        for line_content in lines:
            if qr_name in line_content:
                qr_nameA ,cru_qr_name ,PatternA ,tokkenA, pattern_idA = line_content.split(",")
                return PatternA ,tokkenA, pattern_idA
        
```

```python
qr_name = "Crumbled hash_qr_code[1].png"
```

```python
Pattern ,tokken , pattern_id = get_data(qr_name)
print(Pattern)
print(tokken)
print(pattern_id)
```

```python
qr1 = cv2.imread(qr_name,-1)
plt.imshow(qr1,cmap='gray')
```

```python
t_h,t_w = sf.tokkenning(tokken,qr1.shape[0],qr1.shape[1])
```

```python
if Pattern == "Shuffling":
    for i in range(500):
        qr1 = sf.col_left_right_reverse(qr1,t_w)
    for i in range(500):
        qr1 = sf.row_top_down_reverse(qr1,t_h)   
    plt.imshow(qr1,cmap='gray')
else:
    import Flipping as fl
    for i in range(500):
        qr1 = fl.Left(fl.Up(fl.Right(fl.Down(qr1,t_w),t_h),t_w),t_h)
    plt.imshow(qr1,cmap='gray')
```
